```python
import csv 
reader=csv.reader(open("example.csv",encoding="utf-8"))
single=tuple()
search=1

for i in reader:
    single+=(i[search+4],)

print("班次",search,"的每一個時間:",single)
```

```python
reader=csv.reader(open("example.csv",encoding="utf-8"))
whole=dict()
f=0
for i in reader:
    if(f==0):
        for j in i[5:-1]:
            whole.setdefault(j,tuple())
        f=1
    else:
        c=5
        for j in whole:
            whole[j]+=(i[c],)
            c+=1
print(whole)
```
