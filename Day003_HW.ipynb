```python
import os

files=tuple()
path="Data/Day003"
for i in os.listdir("Data/Day003"):
    if("xml" in i and "CH" in i):
        files+=(i,)
```

```python
import xmltodict
c=0
for i in files:
    with open(path+"/"+i,encoding="utf-8") as f:
        doc=xmltodict.parse(f.read())
    if(doc["cwbopendata"]["dataset"]["locations"]["locationsName"]=="高雄市"):
        for i in doc["cwbopendata"]["dataset"]["locations"]["location"]:
            print(i["locationName"])
            print(i["weatherElement"][0]["time"][0]["dataTime"],i["weatherElement"][0]["time"][0]["elementValue"]["value"],i["weatherElement"][0]["time"][0]["elementValue"]["measures"])
            c+=1
        for i in doc["cwbopendata"]["dataset"]["locations"]["location"]:
            print(i["locationName"])
            for j in i["weatherElement"][0]["time"]:
                print(j["dataTime"],j["elementValue"]["value"],j["elementValue"]["measures"])
            break
        break
print("高雄市有",c,"地區有溫度資料")
```
